```python
df["Daily Revenue (£)"] = df["Customers per Day"] * df["Avg Meal Price (£)"]
```

```python
df["Weekly Revenue (£)"] = df["Daily Revenue (£)"] * 7
```

```python
df["Yearly Revenue (£)"] = df["Daily Revenue (£)"] * 365
```

```python
df["Food Cost (£)"] = df["Daily Revenue (£)"] * df["Food Cost %"]

df["Daily Profit (£)"] = df["Daily Revenue (£)"] - df["Food Cost (£)"]

df["Weekly Profit (£)"] = df["Daily Profit (£)"] * 7

df["Yearly Profit (£)"] = df["Daily Profit (£)"] * 365
```
