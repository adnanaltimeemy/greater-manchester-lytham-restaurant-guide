```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
burger_king_data = {

    "brand": ["Burger King"] * 6,

    "combo_name": [
        "King Combo",
        "Whopper Meal",
        "Chicken Royale Meal",
        "Family Bundle",
        "Student Deal",
        "Lunch Saver Combo"
    ],

    "price": [4.99, 8.99, 7.99, 14.99, 5.99, 6.49],

    # Estimated production cost (40%)
    "cost": [2.00, 3.60, 3.20, 6.00, 2.40, 2.60],

    "customers_per_day": [220, 150, 180, 90, 130, 200],

    "city": ["Manchester"] * 6,

    "demographic": [
        "Students",
        "Workers",
        "Workers",
        "Families",
        "Students",
        "Workers"
    ],

    "age_group": [
        "18-25",
        "26-40",
        "26-40",
        "30-50",
        "18-25",
        "18-40"
    ]
}

bk_df = pd.DataFrame(burger_king_data)
bk_df
```

```python
subway_data = {

    "brand": ["Subway"] * 6,

    "combo_name": [
        "6-inch Meal Deal",
        "Footlong Combo",
        "Student Footlong Deal",
        "Breakfast Combo",
        "Family Sub Pack",
        "Lunch Special"
    ],

    "price": [8.49, 10.29, 7.99, 5.49, 15.99, 6.99],

    "cost": [3.50, 4.20, 3.20, 2.20, 6.40, 2.80],

    "customers_per_day": [200, 140, 160, 90, 80, 190],

    "city": ["Manchester"] * 6,

    "demographic": [
        "Workers",
        "Workers",
        "Students",
        "Workers",
        "Families",
        "Workers"
    ],

    "age_group": [
        "18-40",
        "26-50",
        "18-25",
        "26-40",
        "30-50",
        "18-40"
    ]
}

sub_df = pd.DataFrame(subway_data)
sub_df
```

```python
df = pd.concat([bk_df, sub_df], ignore_index=True)
df
```

```python
df["profit_per_sale"] = df["price"] - df["cost"]

df["customers_per_week"] = df["customers_per_day"] * 7

df["revenue_weekly"] = df["price"] * df["customers_per_week"]

df["profit_weekly"] = df["profit_per_sale"] * df["customers_per_week"]

df["profit_yearly"] = df["profit_weekly"] * 52

df
```

```python
profit_summary = df.groupby("brand")[["profit_weekly","profit_yearly"]].sum()

profit_summary
```

```python
demographic_profit = df.groupby(["brand","demographic"])["profit_yearly"].sum()

demographic_profit
```

```python
plt.figure(figsize=(10,5))

sns.barplot(data=df, x="combo_name", y="profit_weekly", hue="brand")

plt.title("Weekly Profit by Combo")

plt.xticks(rotation=45)

plt.show()
```

```python
stores = 25   # example Manchester

manchester_profit = df.groupby("brand")["profit_yearly"].sum() * stores

manchester_profit
```

```python
bk_uk_profit = df[df["brand"]=="Burger King"]["profit_yearly"].sum() * 570

subway_uk_profit = df[df["brand"]=="Subway"]["profit_yearly"].sum() * 650  # estimate

bk_uk_profit, subway_uk_profit
```

```python
def choose_restaurant(age):

    if age <= 25:
        return np.random.choice(["Subway","Burger King"], p=[0.6,0.4])

    elif age <= 40:
        return np.random.choice(["Burger King","Subway"], p=[0.6,0.4])

    else:
        return np.random.choice(["Subway","Burger King"], p=[0.5,0.5])

ages = np.random.randint(18,60,1000)

choices = [choose_restaurant(age) for age in ages]

pd.Series(choices).value_counts()
```
