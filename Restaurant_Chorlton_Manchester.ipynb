```python
import pandas as pd

# dataset example
data = {

    "restaurant": [
        "Rudy's Pizza Chorlton",
        "Delhi Chaat N Cafe",
        "Turkish Delight",
        "Royal Cod",
        "ZAXX Fried",
        "YARA Restaurant",
        "Double Zero Pizza"
    ],

    "food_type": [
        "Pizza",
        "Indian",
        "Turkish",
        "Fish & Chips",
        "Fried Chicken",
        "Lebanese",
        "Italian"
    ],

    "combo_deal": [
        "Pizza + Drink",
        "Curry + Rice + Drink",
        "Kebab + Fries + Drink",
        "Fish + Chips + Drink",
        "Chicken Box Meal",
        "Mixed Grill Combo",
        "Family Pizza Pack"
    ],

    "combo_price": [15, 12, 14, 10, 9, 18, 25],

    "customers_per_day": [200, 150, 180, 120, 250, 130, 170],

    "customer_type": [
        "Young Professionals",
        "Students",
        "Mixed",
        "Families",
        "Students",
        "Families",
        "Mixed"
    ],

    "age_group": [
        "18-40",
        "18-30",
        "18-50",
        "30-60",
        "16-30",
        "25-55",
        "18-50"
    ]
}

df = pd.DataFrame(data)

df
```

```python
# daily revenue
df["daily_revenue"] = df["combo_price"] * df["customers_per_day"]

# weekly revenue
df["weekly_revenue"] = df["daily_revenue"] * 7

# yearly revenue
df["yearly_revenue"] = df["daily_revenue"] * 365

# profit margins
df["weekly_profit"] = df["weekly_revenue"] * 0.6
df["yearly_profit"] = df["yearly_revenue"] * 0.6

df
```

```python
uk_chains = pd.DataFrame({

    "restaurant": ["McDonalds", "KFC", "Greggs", "Nandos", "Wagamama"],

    "combo_price": [7, 8, 4, 15, 18],

    "customers_per_day": [800, 600, 500, 400, 350]
})

uk_chains["daily_revenue"] = uk_chains["combo_price"] * uk_chains["customers_per_day"]
uk_chains["yearly_profit"] = uk_chains["daily_revenue"] * 365 * 0.6

uk_chains
```

```python
def customer_demographic(price):

    if price <= 6:
        return "Low income, Students"

    elif price <= 12:
        return "Middle income, workers"

    else:
        return "High income professionals"


df["demographic"] = df["combo_price"].apply(customer_demographic)

df
```

```python
import matplotlib.pyplot as plt

plt.figure()

plt.bar(df["restaurant"], df["yearly_profit"])

plt.title("Yearly Profit by Restaurant")

plt.xticks(rotation=45)

plt.show()
```
