```python
sales_data = {
    "Brand": ["Wingstop", "Dairy Queen"],
    "Customers_per_day": [400, 300],
    "Average_spend": [11.50, 7.50]
}

df_sales = pd.DataFrame(sales_data)

df_sales["Daily_Revenue"] = df_sales["Customers_per_day"] * df_sales["Average_spend"]

df_sales["Weekly_Revenue"] = df_sales["Daily_Revenue"] * 7

df_sales["Yearly_Revenue"] = df_sales["Daily_Revenue"] * 365

df_sales
```

```python
df_sales["Profit_margin"] = [0.25, 0.20]

df_sales["Weekly_Profit"] = df_sales["Weekly_Revenue"] * df_sales["Profit_margin"]

df_sales["Yearly_Profit"] = df_sales["Yearly_Revenue"] * df_sales["Profit_margin"]

df_sales
```

```python
combo_data = {
    "Brand": ["Wingstop", "Wingstop", "Dairy Queen", "Dairy Queen"],
    "Combo": [
        "6 Wing Combo",
        "Family Pack",
        "Blizzard Combo",
        "Family Ice Cream Pack"
    ],

    "Price": [9.99, 24.99, 6.99, 14.99],

    "Profit": [5.99, 14.99, 4.49, 8.99]
}

df_combo = pd.DataFrame(combo_data)
df_combo
```

```python
plt.figure(figsize=(10,5))

sns.barplot(
    x="Brand",
    y="Yearly_Profit",
    data=df_sales
)

plt.title("Wingstop vs Dairy Queen Yearly Profit UK")
plt.show()
```
