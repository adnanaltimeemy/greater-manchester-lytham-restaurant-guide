```python
# STEP 1: Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
# STEP 2: Create restaurant dataset

data = {

    "Restaurant_Name": [
        "Deansgate Grill",
        "Manchester Combo House",
        "Spinningfields Kitchen",
        "Deansgate Express",
        "Northern Food Bar",
        "Urban Manchester Eats"
    ],

    "Location": [
        "Deansgate",
        "Deansgate",
        "Spinningfields",
        "Deansgate Locks",
        "City Centre",
        "Deansgate"
    ],

    "Combo_Name": [
        "Student Combo",
        "Family Combo",
        "Premium Combo",
        "Lunch Combo",
        "Budget Combo",
        "Business Combo"
    ],

    "Combo_Price": [
        8.99,
        24.99,
        29.99,
        10.99,
        6.99,
        19.99
    ],

    "Cost_Per_Combo": [
        3.50,
        12.00,
        14.00,
        4.50,
        2.50,
        9.00
    ],

    "Customers_Per_Day": [
        150,
        80,
        60,
        120,
        200,
        90
    ],

    "Customer_Type": [
        "Students",
        "Families",
        "Professionals",
        "Workers",
        "Low Income",
        "Business Professionals"
    ],

    "Age_Group": [
        "18-25",
        "30-50",
        "25-55",
        "18-40",
        "18-60",
        "25-60"
    ]
}

df = pd.DataFrame(data)

df
```

```python
# STEP 3: Profit per combo

df["Profit_Per_Combo"] = df["Combo_Price"] - df["Cost_Per_Combo"]

df
```

```python
# STEP 4: Daily profit

df["Daily_Profit"] = df["Profit_Per_Combo"] * df["Customers_Per_Day"]

df
```

```python
# STEP 5: Weekly profit

df["Weekly_Profit"] = df["Daily_Profit"] * 7

df
```

```python
# STEP 6: Yearly profit

df["Yearly_Profit"] = df["Daily_Profit"] * 365

df
```

```python
# STEP 7: Revenue calculations

df["Daily_Revenue"] = df["Combo_Price"] * df["Customers_Per_Day"]

df["Weekly_Revenue"] = df["Daily_Revenue"] * 7

df["Yearly_Revenue"] = df["Daily_Revenue"] * 365

df
```

```python
# STEP 8: Total profits

total_daily_profit = df["Daily_Profit"].sum()
total_weekly_profit = df["Weekly_Profit"].sum()
total_yearly_profit = df["Yearly_Profit"].sum()

print("Total Daily Profit:", total_daily_profit)
print("Total Weekly Profit:", total_weekly_profit)
print("Total Yearly Profit:", total_yearly_profit)
```

```python
# STEP 9: Customers by type

customer_counts = df.groupby("Customer_Type")["Customers_Per_Day"].sum()

customer_counts.plot(kind="bar", title="Customers by Type")

plt.show()
```

```python
# STEP 10: Profit comparison

plt.figure(figsize=(10,5))

sns.barplot(
    x="Restaurant_Name",
    y="Yearly_Profit",
    data=df
)

plt.title("Yearly Profit by Restaurant")

plt.xticks(rotation=45)

plt.show()
```

```python
# STEP 11: Best combo

best_combo = df.sort_values("Yearly_Profit", ascending=False)

best_combo.head(1)
```

```python
# STEP 12: Save dataset

df.to_csv("deansgate_restaurant_dataset.csv", index=False)
```
