```python
# Step 2: Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from faker import Faker
fake = Faker()

np.random.seed(42)
```

```python
# Step 3: Create Manchester halal restaurant dataset

restaurants = [

    {"name":"MyLahore Manchester", "area":"Rusholme", "type":"Asian Fusion", "avg_meal_price":12, "combo_price":18, "customers_per_day":320, "demographic":"Students"},
    
    {"name":"Toro's Steakhouse Manchester", "area":"Rusholme", "type":"Steakhouse", "avg_meal_price":22, "combo_price":35, "customers_per_day":180, "demographic":"Families"},
    
    {"name":"Gaza Fried Chicken", "area":"Upper Brook Street", "type":"Fast Food", "avg_meal_price":8, "combo_price":14, "customers_per_day":450, "demographic":"Students"},
    
    {"name":"Zouk Tea Bar & Grill", "area":"City Centre", "type":"Indian Grill", "avg_meal_price":18, "combo_price":28, "customers_per_day":240, "demographic":"Professionals"},
    
    {"name":"M.A.K Halal Manchester", "area":"Cheetham Hill", "type":"Burgers", "avg_meal_price":9, "combo_price":15, "customers_per_day":400, "demographic":"Young Adults"},
    
    {"name":"Safire Restaurant", "area":"Oxford Road", "type":"Indian", "avg_meal_price":15, "combo_price":25, "customers_per_day":220, "demographic":"Students"},
    
    {"name":"Al Jazeera Restaurant", "area":"Rusholme", "type":"Middle Eastern", "avg_meal_price":11, "combo_price":18, "customers_per_day":310, "demographic":"Families"},
    
    {"name":"Etci Mehmet Steakhouse", "area":"City Centre", "type":"Turkish Steakhouse", "avg_meal_price":28, "combo_price":45, "customers_per_day":160, "demographic":"High Income"}
]

df = pd.DataFrame(restaurants)

df
```

```python
# Step 4: Revenue calculations

df["daily_revenue"] = df["customers_per_day"] * df["avg_meal_price"]

df["weekly_revenue"] = df["daily_revenue"] * 7

df["yearly_revenue"] = df["weekly_revenue"] * 52

df
```

```python
# Step 5: Profit calculations

df["operating_cost"] = df["yearly_revenue"] * 0.65

df["profit"] = df["yearly_revenue"] * 0.35

df["weekly_profit"] = df["profit"] / 52

df
```

```python
# Step 6: Profit ranking

df_sorted = df.sort_values(by="profit", ascending=False)

df_sorted[["name", "yearly_revenue", "profit"]]
```

```python
# Step 7: Profit visualization

plt.figure(figsize=(12,6))

sns.barplot(
    x="profit",
    y="name",
    data=df_sorted
)

plt.title("Yearly Profit by Manchester Halal Restaurants")
plt.show()
```

```python
# Step 8: Demographic grouping

demo = df.groupby("demographic").sum(numeric_only=True)

demo[["customers_per_day", "profit"]]
```

```python
# Step 9: Combo deal simulation

df["customers_with_combo"] = df["customers_per_day"] * 1.2

df["combo_revenue_daily"] = df["customers_with_combo"] * df["combo_price"]

df["combo_revenue_yearly"] = df["combo_revenue_daily"] * 365

df["combo_profit"] = df["combo_revenue_yearly"] * 0.35

df[["name", "combo_profit"]]
```

```python
# Step 10: Save dataset

df.to_csv("manchester_halal_restaurants_profit_dataset.csv", index=False)
```
